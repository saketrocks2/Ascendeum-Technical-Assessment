```python
n=6
top=0
bottom=n-1
left=0
right=n-1
num=1
res=[[0]*n for i in range(n)]
while top<=bottom and left<=right:
    for i in range(left,right+1):
        res[top][i]=num
        num+=1
    top+=1    
    for j in range(top,bottom+1):
        res[j][right]=num 
        num+=1
    right-=1
    if left<=right:
        for i in  range(right,left-1,-1):
            res[bottom][i]=num
            num+=1
        bottom-=1
    if top<=bottom:
        for i in range(bottom,top-1,-1):
            res[i][left]=num
            num+=1
        left+=1
print(res)                           
for i in range(n):
    for j in range(n):
        print(res[i][j],end=" ")
    print()    
```
